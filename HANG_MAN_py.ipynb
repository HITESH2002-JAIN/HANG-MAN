```
import time
import random
hangmanimages=['''
 +========+                         
 |        |
          | 
          |
          | 
 _________|''','''    
 +========+                        
 |        |
 0        |
          |
          |
 _________|''','''
 +========+                        
 |        |
 0        |
 |        |
          |
 _________|''','''
  +========+                        
  |        |
  0        |
 /|        |
           |
  _________|''','''
  +========+                        
  |        |
  0        |
 /|\       |
           |
  _________|''','''
  +========+                        
  |        |
  0        |
 /|\       |
 /         |
  _________|''','''
  +========+                        
  |        |
  0        |
 /|\       |
 / \       |
  _________|'''] ## Different types of hangman images using special character and symbols
K='Y'
while(K=='Y'): ## Loop for replaying the game
    
    alphabet='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    Words=["PROGRAMMING","PYTHON","CODING","DEVELOPER","BANDWIDTH","ANALOG"
           "COMPUTER","LAPTOP","JAVASCRIPT","KEYBOARD","PHONE","TELEVISION",
           "MICROSOFT","LINUX","HACKING","JAVA","GRAPHICS","HARDWARE","SOFTWARE",
           "PRINTER","MODEM","MONITOR","HARDDISK","LAPTOP","MOUSE"] ## list of words for the secret word
    word=random.choice(Words)
    
    print("LET'S PLAY HANG-MAN")
    time.sleep(0.5)
    print("\n")
    name=input("What is your name? ")
    time.sleep(0.2)
    print("\n")
    print("Hello",name,",Be Ready For Guessing The Secret Word" )
    time.sleep(1)
    print("\n")
    
    print("The secret word has",len(word),"letters")
    time.sleep(0.5)
    print("HINT - The word is related to technology and computing")
    error=0
    guesses=[] ## list to keep a record of all guessed alphabet
    guessed=[] ## list to keep a record of correct alphabet guesses

    while(error<7)and(len(guessed)!=len(set(word))):
        time.sleep(0.8)
        guess=input("GUESS A LETTER FOR THE SECRET WORD ")
        guess=guess.upper()
        
        if(len(guess)==1):
            
            if(guess not in alphabet):## condition to check wether the input is correct or not 
                print("You have not entered a letter")
                continue
            
            elif(guess in guessed) or (guess in guesses):## condition to remind the user wether the alphabet is previously guessed or not 
                print("You have already guessed that letter before.")
                continue
            
            elif(guess not in word):## condition for incorrect guess.
                print("Sorry the letter is not a part of the word :(")
                time.sleep(0.2)
                print(hangmanimages[error])

                time.sleep(0.1)
                print("You have ",6-error,"chances left to guess the word" )
                error=error+1
                continue
            
            elif(guess in word):## condition for correct guess
                print("Well Done,the letter exist in the word!")
                time.sleep(0.2)
                guessed.append(guess)
                
                i=0
                while(i<len(word)):
                    
                    if(word[i] in guessed):
                        print(word[i],end='')
                    
                    else:
                        print("*",end='')
                    i=i+1
                print("\t")
                continue
                
            else:
                print("Sorry,The game has crashed!! Please try restarting it")
                continue
        
        else:
            print("You have entered more than one character while guessing")
            continue

    if(error==7):##condition when the user has reached the maximum allowed errors
        time.sleep(0.2)
        print("You are dead,the correct word is" ,word)
        time.sleep(0.5)
        print("Better luck next time!!")
        
    
    elif(len(guessed)==len(set(word))):##condition when the user guessed the correct word
        time.sleep(0.2)
        print("Congratulations,You have correctly guessed the secret word")
    
    k=input("Do you want to play the game again? Y/N ")##To check wether the user want to play the game again or not 
    K=k.upper()
print("Thank you, for playing the game")


```
